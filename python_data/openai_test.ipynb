```python
import openai
import config
```

```python
openai.api_key = config.openai_key
openai.organization = config.openai_org
```

```python
def generate_response(prompt):
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=60,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response.choices[0].text.strip()

prompt = "What is the capital of France?"
response = generate_response(prompt)
print(response)
```

```python
generate_response("What's the type of currency used in yemen?")
```

```python
generate_response("How many seashells on the seashore did she sell?")
```

```python
generate_response("If a tree falls in the forest and no one hears it, does it make a sound?")
```
